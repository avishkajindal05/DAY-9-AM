Exact Prompt: Write a Python function that finds all pairs in a list that sum to a target number using list comprehensions.                                                                                                                       Exact output:

```python
def find_pairs(lst, target):
    pairs = [(lst[i], lst[j]) 
             for i in range(len(lst)) 
             for j in range(i + 1, len(lst)) 
             if lst[i] + lst[j] == target]
    
    return pairs
```

```python
nums = [1, 2, 3, 4, 5]
target = 6

print(find_pairs(nums, target))
```

```python
nums = [1, 1, 1]
target = 2
print(find_pairs(nums, target))
```

Test 2: ([1, 1, 1], target=2)
Output: [(1, 1), (1, 1), (1, 1)]
❌ Incorrect — returns 3 identical pairs instead of just 1.

Issue 1: Duplicate Pairs with Repeated Values
The AI solution uses index-based iteration, so when the same value appears multiple times, it generates a pair for every combination of indices, not every unique value pair.
For [1, 1, 1], the pairs (i=0,j=1), (i=0,j=2), and (i=1,j=2) all produce (1, 1) — three duplicates of the same logical pair.
Issue 2: O(n²) Time Complexity
The nested loop approach checks every possible index pair, giving O(n²) time and O(n) space (for storing results). For large lists, this becomes a significant bottleneck.

```python
def find_pairs_improved(lst, target):
    seen = set()
    pairs = []

    for num in lst:
        complement = target - num
        if complement in seen:
            pair = (min(num, complement), max(num, complement))
            if pair not in pairs:   # deduplicate value-level pairs
                pairs.append(pair)
        seen.add(num)

    return pairs
```

```python
nums = [1, 1, 1]
target = 2
print(find_pairs_improved(nums, target))
```

The AI-generated code works correctly for lists with all unique elements, but fails on repeated values by returning duplicate pairs. The O(n²) complexity also makes it unsuitable for large inputs. The improved solution fixes both issues: it uses a hash set for O(n) performance and explicitly deduplicates at the value level to handle repeated elements correctly.
